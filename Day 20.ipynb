```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

def generate_wordcloud(text, image_path):
    """
    Generates a WordCloud from the given text and saves it as an image.
    
    Args:
        text (str): The input text for generating the WordCloud.
        image_path (str): The path where the WordCloud image will be saved.
    """
    # Generate the WordCloud
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
    
    # Display the WordCloud using matplotlib
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')  # Hide the axes
    plt.show()
    
    # Save the WordCloud as an image
    wordcloud.to_file(image_path)
    print(f"WordCloud image saved at {image_path}")

# Test the function with the given text
text = 'data science machine learning artificial intelligence'
image_path = 'wordcloud_image.png'
generate_wordcloud(text, image_path)
```
